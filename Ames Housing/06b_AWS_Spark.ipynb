# Pasos realizados en Amazon Web Services

A continuación se detallan los pasos seguidos para ello:

Se accede a **AWS Management Console** al servicio **EMR (ElasticMapReduce**, a la opción **"Create Cluster"**.

Se configuran las siguientes opciones:

* ***Configuración de Software***: 
    - **Spark_ Spark 2.4.3 on Hadoop 2.8.5 YARN wigh Ganglia 3.7.2 and Zeppelin 0.8.1.**

* ***Configuración _Hardware***:
    - **m5.xlarge**: máquinas de propósito general de quinta generación (4 CPUs con 16 GB de RAM)
    - **número de instancias**: 3 (1 máster y 2 esclavos)
    



Se ha creado el clúster con la configuración definida:

Se observa que las máquinas se crean en la ***zona eu-west-3*** (París)



Se comprueba el correcto acceso al clúster, realizando una conexión **ssh** al máster 

Posteriormente se accede al bucket creado, que nos servirá para almacenar los datasets necesarios en la ingesta de datos.

A continuación se muestra la misma pantalla, con los datos **train** y **test** que se van a utilizar, ya subidos al clúster.
